```python
import pandas as pd
```

```python
# Reading our data into a data frame
df = pd.read_stata('20660-0001-Data.dta')
```

```python
# getting statistical summary of our data frame 
df.describe()
```

```python
# selecting the features that are important to us and creating a new data frame
new_df = df[['County', 'StateName', 'CountyName', 'Division', 'Region', 'CBSA', 'CBSA_Status', 'MetroDiv', 'Latitude',
   'Longitude', 'LandArea', 'WaterArea', 'PopBase00', 'Deaths01', 'MedianAge05', 'PctWhite05', 'PctBlack05', 'UnempRate05',
    'EconType04', 'HouseStrs04', 'LowEduc04', 'LowEmp04', 'Retirement04', 'RuralUrban03', 'ClimateRegion', 'PropTaxPerCap02',
    'LocTaxPerCap02', 'CrimeRate04']]
```

```python
new_df.info()
```

```python
# pearson correlation of the numeric variables
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (12,6))
sns.heatmap(new_df.corr(numeric_only = True), annot = True)
```

```python
# filling the NAs in the crime rate column using linear regression
from sklearn.linear_model import LinearRegression
```

```python
df_dropna = new_df.dropna(subset = ['CrimeRate04', 'PopBase00', 'PctWhite05', 'PctBlack05', 'Deaths01'])
```

```python
# Define X and y variables
X = df_dropna[['PopBase00', 'PctWhite05', 'PctBlack05', 'Deaths01']]
y = df_dropna['CrimeRate04']
```

```python
# fit the linear regression model
model = LinearRegression()
model.fit(X,y)
```

```python
# Now, predicting the crime rates for rows with missing values
X_missing = new_df[new_df['CrimeRate04'].isna()][['PopBase00', 'PctWhite05', 'PctBlack05', 'Deaths01']]

predicted_crime_rate = model.predict(X_missing)
```

```python
# Fillin the NAs in crime rate with the predicted values
new_df.loc[new_df['CrimeRate04'].isna(), 'CrimerRate'] = predicted_crime_rate
```

```python
new_df[new_df['CrimeRate04'].isna()]
```

```python
# lets replace the NaNs in the crime rate column with the predicted values
new_df['CrimeRate04'] = new_df['CrimeRate04'].fillna(new_df['CrimerRate'])
```

```python
new_df.info()
```

```python
# dropping the derived column and dropping all rows with missing values
new_df.drop('CrimerRate', axis = 1, inplace = True)
new_df.dropna(inplace = True)
new_df.info()
```

```python
# Performing data augmentation to increase the size of our dataset and improve the performance and generalization of our model

# Duplicating our dataframe with modifications
duplicated_df = new_df.sample(n = 8560, replace = True)
```

```python
# Adding noise to the dataset
import numpy as np

def add_noise(x):
    return x + np.random.normal(0, 0.1, len(x))

augmented_df = new_df.copy()
numerical_features = ['Latitude', 'Longitude', 'LandArea', 'WaterArea', 'PopBase00', 'Deaths01', 'MedianAge05',
                      'PctWhite05', 'PctBlack05', 'UnempRate05', 'PropTaxPerCap02', 'LocTaxPerCap02', 'CrimeRate04']

for feature in numerical_features:
    augmented_df[feature] = add_noise(augmented_df[feature])
```

```python
# shuffling rows
shuffled_df = new_df.sample(frac=1).reset_index(drop=True)
```

```python
# concatenating the augmented dataframes
final_df = pd.concat([new_df, duplicated_df, augmented_df, shuffled_df], ignore_index = True)

# checking the size of the final dataframe
print('Size of the final DataFrame: ', final_df.shape)
```

```python
# splitting the data into train and test sets
from sklearn.model_selection import train_test_split
```

```python
X = final_df.drop('CrimeRate', axis = 1)
y = final_df['CrimeRate']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# training and evaluating linear regression model
from sklearn.linear_model import LinearRegression
LR_model = LinearRegression()
```

```python
LR_model.fit(X_train, y_train)
```

```python
# getting model's prediction
LR_prediction = LR_model.predict(X_test)
```

```python
# visualizing the model's performance on new data
sns.scatterplot(x = LR_prediction, y = y_test)
plt.title("Linear Regression Model Prediction against y_test" )
plt.xlabel('Prediction')
plt.ylabel('y test')
```

```python
# model evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, LR_prediction)
mse = mean_squared_error(y_test, LR_prediction)
rmse = np.sqrt(mse)

print(f'mean absolute error: {round(mae, 2)}')
print(f'mean squared error: {round(mse, 2)}')
print(f'Root mean squared error: {round(rmse, 2)}')
```

```python
# lets scale our variables, optimize our model parameters and check for a different result still using SVR
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# using grid search to search for the best hyperparameters
from sklearn.model_selection import GridSearchCV

param_grid = {
    'kernel':['rbf'],
    'C':[0.1, 1, 10, 100],
    'gamma':[0.1, 1, 10, 100],
    'epsilon':[0.1, 0.01, 0.001]
}
grid_search = GridSearchCV(SVR(), param_grid, cv = 5,verbose = 4)
```

```python
grid_search.fit(X_train_scaled, y_train)
```

```python
# getting the best parameters
best_params = grid_search.best_params_

print(f'Best parameters: {best_params}')
```

```python
# using the best params to create a new SVR model
best_svr = SVR(**best_params)

best_svr.fit(X_train_scaled, y_train)
```

```python
best_svr_pred = best_svr.predict(X_test_scaled)
```

```python
sns.scatterplot(x = best_svr_pred, y = y_test)
plt.title("Tuned SVR Model Prediction against y_test" )
plt.xlabel('Prediction')
plt.ylabel('y test')
```

```python
mae = mean_absolute_error(y_test, best_svr_pred)
mse = mean_squared_error(y_test, best_svr_pred)
rmse = np.sqrt(mse)

print(f'mean absolute error: {round(mae, 2)}')
print(f'mean squared error: {round(mse, 2)}')
print(f'Root mean squared error: {round(rmse, 2)}')
```

```python
# using deep learning sequential model
from keras.models import Sequential
from keras.layers import Dense
```

```python
# defining X and y variables and performing train-test split
X = final_df.drop('CrimeRate', axis = 1).values
y = final_df['CrimeRate'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# using minmax scaler to scale the independent variables
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
# instantiating the model and defining its structure and properties
model = Sequential()

model.add(Dense(42, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(42, activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# introducing early stopping callback to our model to monitor validation loss and stop training when it stops improving
from keras.callbacks import EarlyStopping
```

```python
early_stopping = EarlyStopping(monitor = 'val_loss', patience = 5, restore_best_weights = True)
```

```python
# training the model
model.fit(x=X_train, y=y_train, epochs=400, batch_size=128, validation_data=(X_test, y_test), callbacks=[early_stopping])
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
# visualizing validation loss
losses.plot()
```

```python
seq_prediction = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, seq_prediction)
mse = mean_squared_error(y_test, seq_prediction)
rmse = np.sqrt(mse)

print(f'mean absolute error: {round(mae, 2)}')
print(f'mean squared error: {round(mse, 2)}')
print(f'Root mean squared error: {round(rmse, 2)}')
```
